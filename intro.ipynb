Converting group.txt into pandas dataframe

```python
import re #regular expression
import pandas as pd
```

```python
f=open('group_chat.txt','r',encoding='utf-8')
```

```python
data=f.read()
```

```python
# we have converted all data as string
print(type(data))
```

```python
# print(data)
```

```python
# using regex
pattern='\d{1,2}/\d{1,2}/\d{2},\s\d{1,2}:\d{2}\s-\s'
```

```python
messages=re.split(pattern,data)[1:] #first message empty aa rha tha esiliye 1 ke baad wale lena
# messages
len(messages)
```

```python
dates=re.findall(pattern,data)
dates
```

```python
df=pd.DataFrame({'user_message':messages,'message_date':dates})

# converting message date type
df['message_date']=pd.to_datetime(df['message_date'],format='%m/%d/%y, %H:%M - ')
df.rename(columns={'message_date':'date'},inplace=True)

df.head()
```

```python
# upar wale me
# user name and txt are in same colm so seperating them
users=[]
messages=[]

for message in df['user_message']:
    entry=re.split('([\w\W]+?):\s',message)
    if entry[1:]:
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('group_notification')
        messages.append(entry[0])

df['user']=users
df['message']=messages
df.drop(columns=['user_message'],inplace=True)

df.head()
```

```python
# making new colm of year
df['year']=df['date'].dt.year
```

```python
df.head()
```

```python
df['month']=df['date'].dt.month_name()
df['day']=df['date'].dt.day
df['hour']=df['date'].dt.hour
df['minute']=df['date'].dt.minute
```
